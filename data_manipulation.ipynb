```python
import pandas as pd
from pathlib import Path
import re
import string
```

```python
data = pd.read_csv("PoetryFoundationData.csv")
data
```

```python
titles = data['Title']
```

```python
poems = data["Poem"]
```

```python
#Trying to remove non-English poems to improve training 
non_english_idx = []
non_english_words = {'haz','tengo','que','quiero','mas','pas','tres','aussi','comme','je','se','el','la','le','se','te'}
for i in range(len(poems)):
    lines = poems[i].translate(''.maketrans('', '', string.punctuation.replace('-.','')+'☽❖')).replace('—', ' ').replace('.', ' ').replace('\t', '')
    poem_words = list(set(lines.split(' ')))
    for word in poem_words:
        if word in non_english_words:
            non_english_idx.append(i)
```

```python
#Poems to txt files
data_dir = Path("Poems")
if data_dir.is_dir() == False:
    data_dir.mkdir()
for i in range(len(poems)):
    with open("Poems/"+str(i)+".txt", 'w') as f:
        lines = poems[i].translate(''.maketrans('', '', string.punctuation.replace('-.','')+'☽❖')).replace('—', ' ').replace('.', ' ').replace('\t', '')
        f.write(lines+'\n')
```

```python
assert len(poems) == len(titles), "poems and titles lengths mismatch"
data_dir = Path("Titles")
if data_dir.is_dir() == False:
    data_dir.mkdir()
for i in range(len(titles)):
    with open("Titles/"+str(i)+".txt", 'w') as f:
        lines = poems[i].translate(''.maketrans('', '', string.punctuation.replace('-.','')+'☽❖')).replace('—', ' ').replace('.', ' ').replace('\t', '')
        f.write(lines+'\n')
```

```python
data_dir = Path("English_Poems")
non_english_words = {'haz','tengo','que','quiero','mas','pas','tres','aussi','comme','je','se','el','la','le','se','te'}
if data_dir.is_dir() == False:
    data_dir.mkdir()
poem_list = []
for i in range(len(poems)):
    with open("Poems/"+str(i)+".txt", 'r') as f:
        poem_output=''
        txt_string = f.readlines()
        poem = []
        for i in range(len(txt_string)):
            if txt_string[i] == '\n':
                txt_string[i] = ''
                continue
            if txt_string[i].count('\n') == 1 and len(txt_string[i]) > 2:
                txt_string[i] = txt_string[i][:-1] + ' \n '
            poem.append(txt_string[i])
        poem = [[y for y in sentence.split(" ")] for sentence in poem]
        for i in range(len(poem)):
            temp = ''
            for word in poem[i]:
                if word in non_english_words:
                    poem[i].pop()
                    temp = ''
                    break
                temp += ' ' + word
            poem_output += '' + temp
        poem_list.append(poem_output)
for i in range(len(poem_list)):
    with open("English_Poems/"+str(i)+".txt", 'w') as f:
        f.write(poem_list[i]+'\n')
```
